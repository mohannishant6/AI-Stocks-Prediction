```python
import pandas as pd
import numpy as np
import yfinance as yf
import random

dfc = pd.read_csv("Stock codes.csv",sep=",")
l = dfc.set_index('Symbol').to_dict()['Description']
df = pd.DataFrame()
lr = random.sample(list(l),100)

for i in lr:
    dftemp = yf.download(i, 
                      start='2010-01-01', 
                      end='2020-02-29', 
                      progress=False,
                     interval= '1d')
    dftemp['Stock Symbol'] = i
    dftemp['Stock name'] = l[i]
    df = df.append(dftemp)
    
df2 = df[['Stock Symbol','Close']]
df2.to_csv("main_data.csv")
```
